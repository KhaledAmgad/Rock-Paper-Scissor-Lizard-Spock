```python
import numpy as np
import cv2
import time
import random


cap = cv2.VideoCapture("http://127.0.0.1:4747/mjpegfeed")
cap.set(cv2.CAP_PROP_FRAME_HEIGHT,480)
cap.set(cv2.CAP_PROP_FRAME_WIDTH,640)
hands=cv2.imread("hands.png")
hands=cv2.resize(hands,(640,300))
rules=cv2.imread("rules.png")
rules=cv2.resize(rules,(500,430))


scissors=cv2.imread("scissor.png")
scissors=cv2.resize(scissors,(100,100))


rock=cv2.imread("rock.png")
rock=cv2.resize(rock,(100,100))


paper=cv2.imread("paper.png")
paper=cv2.resize(paper,(100,100))


spock=cv2.imread("spock.png")
spock=cv2.resize(spock,(100,100))


lizard=cv2.imread("lizard.png")
lizard=cv2.resize(lizard,(100,100))


gestureImgs = {}

gestureImgs["scissors"] = scissors
gestureImgs["rock"] = rock
gestureImgs["paper"] = paper
gestureImgs["spock"] = spock
gestureImgs["lizard"] = lizard
gestureImgs[""] = newFrame=np.zeros((100,100,3), np.uint8)









start =time.time()
fCount=0
frameRate=0
# font 
font = cv2.FONT_HERSHEY_SIMPLEX 

# org 
org = (50, 50) 

# org 
orgState = (50, 450) 

# org 
orgTitle = (700, 480) 
# org 
orgGesture = (700, 620) 

# org 
orgScore = (700, 720) 

# fontScale 
fontScale = 1

# Blue color in BGR 
color = (255, 255, 255) 

# Blue color in BGR 
red = (0, 255, 255) 

# Line thickness of 2 px 
thickness = 2

# Using cv2.putText() method 

xOffset=500
yOffset=300


state="Wait Start Or End"

gestures=["rock", "paper", "scissors", "lizard", "spock"]
gesturesUser=""
gesturesCpu=""
whoWinString=""
userScore=0
cpuScore=0


    
def drawFrameRate():
    global fCount,frameRate,start,frame
    fCount=fCount+1
    end =time.time()
    if(end-start>=1): 
        frameRate=fCount
        fCount=0
        start =time.time()
    frame = cv2.putText(frame, str(frameRate)+" FPS", org, font, fontScale, color, thickness, cv2.LINE_AA)
    
    
def appendUI():
    global newFrame,frame,hands,rules,state
    frame = cv2.putText(frame, "State :"+ state+"       "+str(whoWinString), orgState , font, fontScale, color, thickness, cv2.LINE_AA)
    newFrame=np.zeros((frame.shape[0]+yOffset,(frame.shape[1]+xOffset),frame.shape[2]), np.uint8)
    newFrame[0:frame.shape[0],0:frame.shape[1],:]=frame
    newFrame[frame.shape[0]:frame.shape[0]+hands.shape[0]    ,    0:hands.shape[1]   ,:]=hands
    newFrame[   0:rules.shape[0]  ,    frame.shape[1]:frame.shape[1]+rules.shape[1]   ,:]=rules
    newFrame[   rules.shape[0]+100:rules.shape[0]+200,  frame.shape[1]+50:frame.shape[1]+150   ,:]=gestureImgs[gesturesUser]
    newFrame[   rules.shape[0]+100:rules.shape[0]+200,  frame.shape[1]+300:frame.shape[1]+400   ,:]=gestureImgs[gesturesCpu]


    newFrame = cv2.putText(newFrame, "USER     VS     CPU", orgTitle, font, fontScale, red, thickness, cv2.LINE_AA)
    #newFrame = cv2.putText(newFrame, str(gesturesUser)+"           "+str(gesturesCpu), orgGuesture, font, fontScale, red, thickness, cv2.LINE_AA)
    newFrame = cv2.putText(newFrame, str(userScore)+"               "+str(cpuScore), orgScore, font, fontScale, red, thickness, cv2.LINE_AA)


    
def whoWin():
    global gesturesUser,gesturesCpu,userScore,cpuScore,whoWinString
    if(gesturesUser==gesturesCpu):
        whoWinString="Draw"
        
    if(gesturesUser=="rock"):
        if(gesturesCpu=="paper") or (gesturesCpu=="spock"):
            whoWinString="cpu"
        elif(gesturesCpu=="scissors") or (gesturesCpu=="lizard"):
            whoWinString="user"

    elif(gesturesUser=="paper"):
        if(gesturesCpu=="rock") or (gesturesCpu=="spock"):
            whoWinString="user"
        elif(gesturesCpu=="scissors") or (gesturesCpu=="lizard"):
            whoWinString="cpu"
         
    elif(gesturesUser=="scissors"):
        if(gesturesCpu=="rock") or (gesturesCpu=="spock"):
            whoWinString="cpu"
        elif(gesturesCpu=="paper") or (gesturesCpu=="lizard"):
            whoWinString="user"
            
    elif(gesturesUser=="lizard"):
        if(gesturesCpu=="rock") or (gesturesCpu=="scissors"):
            whoWinString="cpu"
        elif(gesturesCpu=="paper") or (gesturesCpu=="spock"):
            whoWinString="user"
            
    elif(gesturesUser=="spock"):
        if(gesturesCpu=="rock") or (gesturesCpu=="scissors"):
            whoWinString="user"
        elif(gesturesCpu=="lizard") or (gesturesCpu=="paper"):
            whoWinString="cpu"        

    if(whoWinString=="user"):
        userScore=userScore+1
    elif(whoWinString=="cpu"):
        cpuScore=cpuScore+1



def main():

    global cap,frame,state,gesturesUser,gesturesCpu,userScore,cpuScore,gestures,whoWinString
    while(cap.isOpened()):

        ret, frame = cap.read()


        drawFrameRate()
        
        

        appendUI()
        


        cv2.imshow('frame', newFrame)
        key=cv2.waitKey(1) & 0xFF
        if key == ord('q'):
            break
            
        if state=="Wait Start Or End":

            if key == ord('w'):
                state="Playing"
                
        if state=="Playing":

            if key == ord('p'):
                state="Pause"
                
            if key == ord('z'):
                gesturesUser="rock"  
                gesturesCpu=gestures[random.randrange(5)]
                whoWin()
                state="Pause"
            
            if key == ord('x'):
                gesturesUser="paper"   
                gesturesCpu=gestures[random.randrange(5)]
                whoWin()
                state="Pause"

            
            if key == ord('c'):
                gesturesUser="scissors"  
                gesturesCpu=gestures[random.randrange(5)]
                whoWin()
                state="Pause"

            
            if key == ord('v'):
                gesturesUser="lizard"     
                gesturesCpu=gestures[random.randrange(5)]
                whoWin()
                state="Pause"
            
            if key == ord('b'):
                gesturesUser="spock"     
                gesturesCpu=gestures[random.randrange(5)]
                whoWin()
                state="Pause"
            
                
                
        if state=="Pause":

            if key == ord('r'):
                state="Playing"
                whoWinString=""
                
            if key == ord('w'):
                state="Playing"
                whoWinString=""
                userScore=0
                cpuScore=0


    cap.release()
    cv2.destroyAllWindows()
    

main()
```
